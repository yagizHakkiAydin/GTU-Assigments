```python

from bs4 import BeautifulSoup

# Load the XML file
with open('wiki_00', 'r') as xml_file:
    xml_content = xml_file.read()

# Parse the XML content
soup = BeautifulSoup(xml_content, 'xml')


text = soup.get_text()
```

```python
import re

def clean_text(input_text):
    # Remove all punctuations except space, question mark, exclamation mark, and period
    cleaned_text = re.sub(r'[^\w\s.!?]', '', input_text)
    
    # Remove other symbols except a-z
    cleaned_text = re.sub(r'[^a-zA-ZığşçöüİĞŞÇÖÜ\s.!?]', '', cleaned_text)
    
    return cleaned_text


cleaned_text = clean_text(text)

cleaned_text = cleaned_text.lower()

# Replace "." with " xxx "
cleaned_text = cleaned_text.replace(".", " yyy ")

# Replace spaces with "yyy"
cleaned_text = cleaned_text.replace(" ", " xxx ")

#print("Cleaned Text:", cleaned_text2)


```

```python
def flatten_syllables(syllables_list):

    result = []

    for i in range( 0 , len(syllables_list) ):
        for j in range( 0 , len(syllables_list[i]) ):
            result.append(syllables[i][j])
    
    return result

flattened_syllables = flatten_syllables(syllables)
print(flattened_syllables)

```

```python
unigram_table = {}

for i in range( len(unique_syllables) ):
    unigram_table[unique_syllables[i]] = 0


count_total_unigram_frequencies = 0
for syllable in flattened_syllables:
    unigram_table[ syllable ] += 1
    count_total_unigram_frequencies += 1

for syllable in flattened_syllables:
    unigram_table[ syllable ] /= count_total_unigram_frequencies
    

```

```python
bigram_table = {}

for i in range( len(unique_syllables) ):
    temp = {}
    for syllable in unique_syllables:
        temp[ syllable ] = 0

    bigram_table[unique_syllables[i]] = temp

```

```python

max_frequency_of_a_pair = 0
count_total_bigram_frequencies = 0

for i in range( 0 , len(flattened_syllables) - 1 ):
    bigram_table[ flattened_syllables[i] ][ flattened_syllables[i+1] ] += 1
    count_total_bigram_frequencies += 1

    if bigram_table[ flattened_syllables[i] ][ flattened_syllables[i+1] ] > max_frequency_of_a_pair:
        max_frequency_of_a_pair = bigram_table[ flattened_syllables[i] ][ flattened_syllables[i+1] ]



```

```python
frequencies_of_frequencies = [0] * max_frequency_of_a_pair + [0]

for i in range( len(unique_syllables) ):
    for j in range( len(unique_syllables) ):
        frequencies_of_frequencies[ bigram_table[unique_syllables[i]][unique_syllables[j]] ] += 1
```

```python

#aapplying good-turing smoothing
for i in range(len(unique_syllables)):
    for j in range(len(unique_syllables)):
        observed_freq = bigram_table[unique_syllables[i]][unique_syllables[j]]
        observed_count = frequencies_of_frequencies[observed_freq]
        
        # Implement Good-Turing formula
        if observed_freq < max_frequency_of_a_pair:
            next_freq = observed_freq + 1
            next_count = frequencies_of_frequencies[next_freq]
            
            # Calculate adjusted frequency using Good-Turing formula
            adjusted_freq = (next_freq * next_count) / observed_count
            
            # Update bigram_table with adjusted frequency
            bigram_table[unique_syllables[i]][unique_syllables[j]] = adjusted_freq / count_total_bigram_frequencies
        else:
            # Handle bigrams with frequencies equal to max_frequency_of_a_pair (no adjustment)
            pass
```

```python
import random

def generate_sequence(unigram_table, length):
    sequence = []
    for _ in range(length):
        # Sample a syllable based on unigram probabilities
        syllable = random.choices(list(unigram_table.keys()), weights=list(unigram_table.values()))[0]
        sequence.append(syllable)
    return sequence

# Example usage:

generated_sequence = generate_sequence(unigram_table, 5)
print("Generated Sequence:", generated_sequence)

```

```python
import random

def generate_random_sentence(n, ngram_table, syllables, num_sentences=5):
    sentences = []

    for _ in range(num_sentences):
        sentence = []
        for _ in range(n):
            if len(sentence) == 0:
                # For the first word, sample from the unigram model
                syllable = random.choices(syllables, weights=list(ngram_table.values()))[0]
            else:
                # For subsequent words, sample from the bigram model based on the last word
                last_word = sentence[-1]
                bigram_probs = {s: ngram_table.get((last_word, s), 1e-10) for s in syllables}
                syllable = random.choices(syllables, weights=list(bigram_probs.values()))[0]
            
            sentence.append(syllable)

        sentences.append(sentence)

    return sentences



syllables = list(unigram_table.keys())

# Generate random sentences for unigram model
random_sentences_unigram = generate_random_sentence(1, unigram_table, syllables)
print("Random Sentences (Unigram Model):", random_sentences_unigram)







def generate_random_sentence_bigram(bigram_table, syllables, num_sentences=5):
    sentences = []

    for _ in range(num_sentences):
        sentence = []
        for _ in range(5):  # Adjust the length of the sentence as needed
            if len(sentence) == 0:
                # For the first word, sample from the unigram model
                syllable = random.choices(syllables, weights=list(bigram_table[syllable].values()))[0]
            else:
                # For subsequent words, sample from the bigram model based on the last word
                last_word = sentence[-1]
                syllable = random.choices(syllables, weights=list(bigram_table[last_word].values()))[0]
            
            sentence.append(syllable)

        sentences.append(sentence)

    return sentences

ies based on your bigram table

# Generate random sentences for bigram model
random_sentences_bigram = generate_random_sentence_bigram(bigram_table, syllables)
print("Random Sentences (Bigram Model):", random_sentences_bigram)

# Generate random sentences for bigram model
random_sentences_bigram = generate_random_sentence(2, bigram_table, syllables)
print("Random Sentences (Bigram Model):", random_sentences_bigram)
```
